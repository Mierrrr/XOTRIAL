```python
from itertools import combinations

INF = float('inf')

def travelling_salesman(graph, start=0):
    n = len(graph)
    dp = [[INF] * n for _ in range(1 << n)]
    dp[1 << start][start] = 0

    for mask in range(1 << n):
        for u in range(n):
            if not (mask & (1 << u)):
                continue
            for v in range(n):
                if mask & (1 << v):
                    continue
                new_mask = mask | (1 << v)
                dp[new_mask][v] = min(dp[new_mask][v], dp[mask][u] + graph[u][v])

    res = INF
    for i in range(n):
        if i != start:
            res = min(res, dp[(1 << n) - 1][i] + graph[i][start])
    return res


# --- DRIVER CODE ---
if __name__ == "__main__":
    graph = [
        [0, 10, 15, 20],
        [10, 0, 35, 25],
        [15, 35, 0, 30],
        [20, 25, 30, 0]
    ]

    print("Distance Matrix:")
    for row in graph:
        print(row)

    min_cost = travelling_salesman(graph, start=0)
    print("\nMinimum cost to visit all cities and return to start:", min_cost)
```
