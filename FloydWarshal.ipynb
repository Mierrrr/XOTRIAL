```python
# Floyd-Warshall Algorithm in Python

INF = 99999  # Infinity (represents no connection)

def floyd_warshall(graph):
    n = len(graph)
    dist = [row[:] for row in graph]  # copy input graph

    # Compute shortest distances
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
    return dist

def print_solution(dist):
    print("Shortest distance matrix:")
    for row in dist:
        print(" ".join(f"{x if x != INF else 'INF':>4}" for x in row))

# --- DRIVER CODE ---
if __name__ == "__main__":
    graph = [
        [0,   5,  INF, 10],
        [INF, 0,   3,  INF],
        [INF, INF, 0,   1],
        [INF, INF, INF, 0]
    ]

    print("Input Graph (Adjacency Matrix):")
    for row in graph:
        print(row)
    print()

    result = floyd_warshall(graph)
    print_solution(result)
```
