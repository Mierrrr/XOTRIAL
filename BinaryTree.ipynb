```python
class Node:
    def __init__(self, key):
        self.key, self.left, self.right = key, None, None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, key):
        def _ins(root, key):
            if not root: return Node(key)
            if key < root.key: root.left = _ins(root.left, key)
            elif key > root.key: root.right = _ins(root.right, key)
            return root
        self.root = _ins(self.root, key)

    def search(self, key):
        cur = self.root
        while cur:
            if cur.key == key: return True
            cur = cur.left if key < cur.key else cur.right
        return False

    def delete(self, key):
        def _del(root, key):
            if not root: return None
            if key < root.key: root.left = _del(root.left, key)
            elif key > root.key: root.right = _del(root.right, key)
            else:
                if not root.left: return root.right
                if not root.right: return root.left
                temp = root.right
                while temp.left: temp = temp.left
                root.key, root.right = temp.key, _del(root.right, temp.key)
            return root
        self.root = _del(self.root, key)

    def display(self):
        def _inorder(root):
            return _inorder(root.left) + [root.key] + _inorder(root.right) if root else []
        print("Inorder:", *(_inorder(self.root) or ["<empty>"]))

# --- Driver Code ---
if __name__ == "__main__":
    bst = BST()
    for v in [50, 30, 70, 20, 40, 60, 80]:
        bst.insert(v)
    bst.display()

    print("Search 40:", "Found " if bst.search(40) else "Not Found ")
    print("Search 25:", "Found " if bst.search(25) else "Not Found ")

    for v in [20, 30, 50]:
        print(f"\nDelete {v}:")
        bst.delete(v)
        bst.display()
```
